9. Visualize the pokemon's growth rate

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Growth Rate'], bins=10, color='skyblue', edgecolor='black') 
plt.title('Histogram of Pokémon Growth Rates')
plt.xlabel('Growth Rate')
plt.ylabel('Frequency')
plt.show()
```

10.Visualize the pokemons weights for every species

```python
plt.plot(df['Weight'])
plt.title('Line Plot of Pokémon Weights')
plt.xlabel('Index')
plt.ylabel('Weight')
plt.grid(True)
plt.show()
```

11.Visualize the distribution of different gender values in the Pokémon dataset. 

```python
gender=df['Gender'].value_counts()
plt.bar(gender.index,gender,color='g')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.title('Distribution of Pokémon Genders')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

12.Visualize the distribution of Pokémon Catch rate 

```python
plt.hist(df['Catch rate'],bins=10,edgecolor='black',color='m')
plt.title('Distribution of Pokémon Catch Rate')
plt.xlabel('Catch Rate')
plt.ylabel('Frequency')
plt.grid(axis='y')  
plt.show()
```

13.Visualize the  top five Catch rate for different pokemon species in the Pokémon dataset. 

```python
sorted_df = df.sort_values(by='Catch rate', ascending=False).head()
plt.figure(figsize=(10, 6))
plt.bar(sorted_df['Species'], sorted_df['Catch rate'], color='lightgreen', edgecolor='black')
plt.xlabel('Species')
plt.ylabel('Catch Rate')
plt.title('Top 5 Species by Catch Rate')
plt.xticks(rotation=45) 
```

14. Count the number of Pokémon by 'Type' and get the top 10


```python
top_types = df['Type'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_types.plot(kind='bar', color='blue', edgecolor='black')
plt.title('Top 10 Pokémon Types by Count')
plt.xlabel('Type')
plt.ylabel('Number of Pokémon')
plt.xticks(rotation=45)
plt.show()
```

15. Plot a pie chart for the top 10 Pokémon types


```python
plt.figure(figsize=(10, 8))
top_types.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Pokémon Types by Proportion')
plt.ylabel('')  # Hide y-axis label for the pie chart
plt.show()
```

16. Finding the correlation between height and weight

```python
heights = df['Height']
weights = df['Weight']
plt.scatter(heights, weights, color='yellow', edgecolor='black')
plt.title('Scatter Plot of Pokémon Height vs. Weight')
plt.xlabel('Height (m)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()
```

17.Finding the Frequencies of different Pokemon abilities

```python
ability_counts = df['Abilities'].value_counts().head(10)
ability_df = ability_counts.reset_index()
ability_df.columns = ['Ability', 'Count']
plt.figure(figsize=(12, 8))
sns.barplot(x='Count', y='Ability', data=ability_df)
plt.title('Frequency of Different Pokémon Abilities')
plt.xlabel('Frequency')
plt.ylabel('Ability')
plt.show()
```

18. Finding the distribution of base friendship for over all data

```python
sns.boxplot( y='Base Friendship', data=df, palette='Set2')
plt.title('Distribution of base friendship for all over data')
plt.ylabel('Base Friendship')
plt.xticks(rotation=90) 
plt.grid(True)
plt.show()
```

19. Plot the top five hit points from all over pokemon's species 

```python
top_5_hp = df.nlargest(5, 'HP')[['Name', 'HP']]
sns.barplot(x='HP', y='Name', data=top_5_hp, palette='viridis')
plt.title('Top 5 Pokémon by HP')
plt.xlabel('HP')
plt.ylabel('')

```

20. Plot the top five attack score from all over pokemon's species 

```python
top_5_attack = df.nlargest(5, 'Attack')[['Name', 'Attack']]
sns.barplot(x='Attack', y='Name', data=top_5_attack, palette='bright')
plt.title('Top 5 Pokémon by Attack score')
plt.xlabel('Attack')
plt.ylabel('')
```

21. Plot the top five defense score from all over pokemon's species 

```python
top_5_defense = df.nlargest(5, 'Defense')[['Name', 'Defense']]
sns.barplot(x='Defense', y='Name', data=top_5_defense, palette='dark')
plt.title('Top 5 Pokémon by Defense')
plt.xlabel('Defense')
plt.ylabel('')
```
