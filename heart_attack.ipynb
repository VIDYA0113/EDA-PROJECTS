7.Visualize how many male and female members are there in the given data set

```python
a=df.groupby('Gender').size()
plt.pie(a,autopct='%0.1f%%',labels=a.index,colors=['pink','lightblue'],explode=[0.04,0])
plt.title("Males and Females from the over all data") 
```

8. Visualise the age distribution from the data 

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=15, kde=True, color='m')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

9. PLot the Age Wise average blood pressure from the given data 

```python
age_wise_bp = df.groupby('Age')['Blood Pressure (mmHg)'].mean()
plt.plot(age_wise_bp.index, age_wise_bp.values, marker='o')
plt.title('Age-wise Average Blood Pressure')
plt.xlabel('Age')
plt.ylabel('Average Blood Pressure (mmHg)')
plt.show()
```

10. plot the boxplot graph from any column from the data

```python
plt.boxplot(df['Blood Pressure (mmHg)'])
plt.title('Boxplot of Blood Pressure (mmHg)')
plt.ylabel('Blood Pressure (mmHg)')
plt.tight_layout()
plt.show()
```

11 .Visualise the Violinplot of cholesterol by Gender

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='Gender', y='Cholesterol (mg/dL)', data=df, showmeans=True)
plt.title('Violin Plot of Cholesterol (mg/dL) by Gender')
plt.xlabel('Gender')
plt.ylabel('Cholesterol (mg/dL)')
plt.show()
```

12. Horizontal bar plot for gender-wise average blood pressure

```python
gender_wise_bp = df.groupby('Gender')['Blood Pressure (mmHg)'].mean()
plt.barh(gender_wise_bp.index, gender_wise_bp.values, color=['skyblue', 'lightgreen'],edgecolor='black')
plt.title('Gender-wise Average Blood Pressure')
plt.xlabel('Average Blood Pressure (mmHg)')
plt.ylabel('Gender')
plt.grid(axis='x')
plt.show()
```

13. Group by Gender and Smoking Status, then unstack to create separate columns for each smoking status and visualize them 

```python
gender_wise_SS = df.groupby(['Gender', 'Smoking Status']).size().unstack()
gender_wise_SS.plot(kind='bar', stacked=True, color=['red', 'orange', 'yellow'],edgecolor='black')
plt.title('Gender-wise Smoking Status')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

14. Create a pairplot for age,Blood pressure and cholesterol

```python
numerical_columns = ['Age', 'Blood Pressure (mmHg)', 'Cholesterol (mg/dL)']
sns.pairplot(df[numerical_columns])
plt.show()
```

15. Group by Chest Pain Type and Gender, then count occurrences and Visualize them

```python
count_cpt = df.groupby('Chest Pain Type')['Gender'].value_counts().unstack()
count_cpt.plot(kind='bar', stacked=True, color=['pink', 'blue'],edgecolor='black',figsize=(10,6))
plt.title('Chest Pain Type by Gender')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

16.Find the correlation for bp and cholesterol by heatmap

```python
corr_matrix = df[['Blood Pressure (mmHg)', 'Cholesterol (mg/dL)']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='viridis', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

17. plot the scatter plot for bp and cholesterol by the gender

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Blood Pressure (mmHg)', y='Cholesterol (mg/dL)', data=df,hue='Gender')
plt.title('Blood Pressure vs. Cholesterol by Gender')
plt.xlabel('Blood Pressure (mmHg)')
plt.ylabel('Cholesterol (mg/dL)')
plt.grid()
plt.show()
```

18. plot the visualization by Chest Pain Type by Treatment using stacked bar plot

```python
chest_pain_treatment = df.groupby(['Chest Pain Type', 'Treatment']).size().unstack()
plt.figure(figsize=(10, 6))
chest_pain_treatment.plot(kind='bar', stacked=True, color=['lightyellow', 'lightskyblue', 'lightgreen','yellow'], edgecolor='black')
plt.title('Chest Pain Type by Treatment')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.legend(title='Treatment')
plt.show()
```
