8. Visualize the residential status column with gender

```python
status=df.groupby('gender')['residential_status'].value_counts().unstack()
print(status)
status.plot(kind='bar',stacked="True",color=('red','yellow'))
```

9.Visualize count of degree level of the students

```python
ug=df['degree_level'].value_counts()
print(ug)
colors=['pink','green']
ug.plot(kind='pie',autopct='%0.1f%%',colors=colors)
plt.legend()
```

10. Visualize the count of ug students by degree major

```python
major=df[df['degree_level']=='Undergraduate'].groupby('degree_major').size()
print(major)
major.plot(kind='bar',color=['r','g','m','orange'])
plt.title('Count of Undergraduate Students by Degree Major')
plt.xlabel('Degree Major')
plt.ylabel('Count')
```

11. Visualize gender distribution by age with stacked

```python
gender_count=df.groupby('age')['gender'].value_counts().unstack()
print(gender_count)
gender_count.plot(kind='bar', color=['red', 'black'],stacked ='True')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Gender Distribution by Age')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.show()
```

12. Visualize the count of the student per university using bar chart

```python
uni=df.groupby('university').size()
uni.plot(kind='bar',color='lightblue')
plt.xlabel("university")
plt.ylabel("count")
plt.title("Count of the student per university")
plt.xticks(rotation=40)
```

13. Visualize the count of the student per university using count plot

```python
uni = df['university'].value_counts()
sns.countplot(x='university', data=df, order=uni.index)
plt.title('Count of Students per University')
plt.xlabel('University')
plt.ylabel('Count')
plt.xticks(rotation=30)
plt.grid(axis='y')
```

15. Plot which degree major has more academic pressure

```python
pre_acc_major=df.groupby('degree_major')['academic_pressure'].size()
pre_acc_major.plot(kind='barh',color='m')
plt.title("academic pressure per degree major")
```

16. visualize which degree major has more academic workload

```python
pre_acc_major=df.groupby('degree_major')['academic_workload '].size()
pre_acc_major.plot(kind='barh',color='c')
plt.title("academic workload per degree major")
```

17.Visualize the different frequencies for stress relief activities

```python
activity_counts = df['stress_relief_activities'].value_counts()
plt.figure(figsize=(15,10))
activity_counts.plot(kind='barh', color='skyblue')
plt.title('Frequency of Stress Relief Activities')
plt.xlabel('Count')
plt.ylabel('Activity')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

 18.Visualize the distribution of responses related to "future insecurity."

```python
insecurity =df['future_insecurity'].value_counts()
plt.pie(insecurity,labels=insecurity.index,autopct="%0.1f%%",startangle=140)
plt.title('Distribution of Future Insecurity Responses')
```

19.Visualize sports engagement levels across universities using a stacked bar chart

```python
engagement_counts = df.groupby('university')['sports_engagement'].value_counts().unstack()
print(engagement_counts)
plt.figure(figsize=(15, 8))
engagement_counts.plot(kind='bar', stacked=True, color=['c','blue','green','black'])  
plt.title('Sports Engagement Levels Across Universities')
plt.xlabel('University')
plt.ylabel('Count')
plt.legend(title='Sports Engagement')
plt.xticks(rotation=45)
```

20.Visualization of Correlation Between Study Satisfaction and Mental Health Indicators

```python
data=df[['study_satisfaction','financial_concerns',
       'social_relationships','depression','anxiety','isolation']]
corr=data.corr()
sns.heatmap(corr, annot=True, linewidths=0.5)
plt.title('Heatmap of Selected Columns')
plt.show()
```
