12. Visualize the total Medals Count for gold,silver,bronze

```python
year=df.groupby(['Year','Medal']).size() 
filtered = year.loc[(slice(None), ['Gold', 'Silver', 'Bronze'])]
medal_counts = filtered.groupby('Medal').sum()
medal_counts.plot(kind='bar', color=['#cd7f32', 'gold', 'silver'])
plt.xlabel('Medal Type')
plt.ylabel('Count')
plt.title('Total Medal Counts for Gold, Silver, and Bronze')
plt.show()
```

14. Find the top cities with the most players

15. Visualize the year wise medal count from the given data 

```python
c=df.groupby('Year')['Medal'].size()
d=c.nlargest(10)
plt.plot(d.index,d.values,marker='<',color='pink',markerfacecolor='black')
plt.xlabel('Year')
plt.ylabel('Number of Medals')
plt.title('Top 10 Years with the Highest Medal Counts')
plt.xticks(rotation=45)
plt.tight_layout()  
plt.show()
```

16. Find and Visualize the top 5 events in the any one year

```python
sep=df[df['Year']==2024]
cou=sep.groupby('Event').size()
pcou=cou.nlargest(5)
pcou_df = pcou.reset_index(name='Count')
sns.barplot(y='Event',x='Count',data=pcou_df)
plt.title("Top 5 events in 2024 ") 
plt.show()
```

20. Find and Visualise Sports wise medal counts

```python
medal_df = df[df['Medal'] != 'No medal']
sport_medal_counts = medal_df.groupby('Sport')['Medal'].count().reset_index()
sport_medal_counts = sport_medal_counts.sort_values(by='Medal', ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x='Sport', y='Medal', data=sport_medal_counts, palette='viridis')
plt.title('Sport-wise Medal Counts')
plt.xlabel('Sport')
plt.ylabel('Number of Medals')
plt.xticks(rotation=90)
plt.show()
```
